```python
import seaborn as sns
import pandas as pd
```

```python
dat = pd.read_csv('Zomato-data-.csv')
dat.rename(columns={'approx_cost(for two people)': 'cost'}, inplace=True)
dat.rename(columns={'listed_in(type)': 'type'}, inplace=True)
dat['rate'] = dat['rate'].str.replace('/5', '').astype(float)
```

```python
dat.head()
```

```python
# distribution plot of rates
sns.displot(dat['rate'], kde=True, bins=20)
```

```python
# Joint plots for joint analysis, joint plot b/w rates and votes
sns.jointplot(x='rate', y='votes', data=dat, kind='hex')
```

```python
# Joint plots for joint analysis, joint plot b/w rates and votes [KDE type]
sns.jointplot(x='rate', y='votes', data=dat, kind='kde')
```

```python
sns.pairplot(dat)
```
